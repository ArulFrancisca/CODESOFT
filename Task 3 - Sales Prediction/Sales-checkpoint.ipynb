```python
import os
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import joblib

DATA_PATH = "sales.csv"
MODEL_SAVE_PATH = "best_sales_model.joblib"
RANDOM_STATE = 42
TEST_SIZE = 0.2
```

```python
EXPECTED_COLS = ["TV", "Radio", "Newspaper", "Sales"]

if not os.path.exists(DATA_PATH):
    raise FileNotFoundError(f"File not found: {DATA_PATH}")

df = pd.read_csv(DATA_PATH)
missing = [c for c in EXPECTED_COLS if c not in df.columns]
if missing:
    raise ValueError(f"Missing columns: {missing}")

df = df[EXPECTED_COLS].dropna().reset_index(drop=True)
print("Shape:", df.shape)
df.head()
```

```python
print("Summary statistics:")
display(df.describe())

print("\nCorrelation:")
display(df.corr()['Sales'].sort_values(ascending=False))

pd.plotting.scatter_matrix(df, figsize=(8,8))
plt.show()
```

```python
X = df[["TV", "Radio", "Newspaper"]]
y = df["Sales"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE
)

scaler = StandardScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Train:", X_train.shape, "Test:", X_test.shape)
```

```python
# Linear Regression
lr = LinearRegression()
lr.fit(X_train, y_train)

# Ridge Regression
ridge = Ridge()
ridge_params = {'alpha':[0.01,0.1,1,10,50,100]}
ridge_cv = GridSearchCV(ridge, ridge_params, cv=5, scoring="neg_mean_squared_error")
ridge_cv.fit(X_train_scaled, y_train)
best_ridge = ridge_cv.best_estimator_

# Lasso Regression
lasso = Lasso(max_iter=10000)
lasso_params = {'alpha':[0.0001,0.001,0.01,0.1,1,10]}
lasso_cv = GridSearchCV(lasso, lasso_params, cv=5, scoring="neg_mean_squared_error")
lasso_cv.fit(X_train_scaled, y_train)
best_lasso = lasso_cv.best_estimator_

print("Linear model trained")
print("Ridge best alpha:", ridge_cv.best_params_)
print("Lasso best alpha:", lasso_cv.best_params_)
```

```python
def metrics(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    return {
        "RMSE": np.sqrt(mse),
        "MAE": mean_absolute_error(y_true, y_pred),
        "R2": r2_score(y_true, y_pred)
    }

pred_lr = lr.predict(X_test)
pred_ridge = best_ridge.predict(X_test_scaled)
pred_lasso = best_lasso.predict(X_test_scaled)

results = pd.DataFrame({
    "Model": ["Linear", "Ridge", "Lasso"],
    "RMSE": [
        metrics(y_test, pred_lr)["RMSE"],
        metrics(y_test, pred_ridge)["RMSE"],
        metrics(y_test, pred_lasso)["RMSE"],
    ],
    "MAE": [
        metrics(y_test, pred_lr)["MAE"],
        metrics(y_test, pred_ridge)["MAE"],
        metrics(y_test, pred_lasso)["MAE"],
    ],
    "R2": [
        metrics(y_test, pred_lr)["R2"],
        metrics(y_test, pred_ridge)["R2"],
        metrics(y_test, pred_lasso)["R2"],
    ]
})

results
```

```python
best_model_name = results.loc[results['RMSE'].idxmin(), 'Model']

if best_model_name == "Linear":
    final_model = lr
    final_scaler = None
elif best_model_name == "Ridge":
    final_model = best_ridge
    final_scaler = scaler
else:
    final_model = best_lasso
    final_scaler = scaler

joblib.dump({
    "model": final_model,
    "scaler": final_scaler,
    "features": ["TV", "Radio", "Newspaper"]
}, MODEL_SAVE_PATH)

print("Best model saved as:", MODEL_SAVE_PATH)
print("Best model:", best_model_name)
```

```python
sample = pd.DataFrame({
    "TV": [150],
    "Radio": [25],
    "Newspaper": [20]
})

if final_scaler:
    sample_scaled = final_scaler.transform(sample)
    prediction = final_model.predict(sample_scaled)
else:
    prediction = final_model.predict(sample)

print("Sample:", sample)
print("Predicted Sales:", float(prediction[0]))
```

```python
 # ---------- Cell 9: Dashboard (2x2) ----------
# Requires: X_test (DataFrame), y_test (Series/array), final_model, final_scaler (or pred_* variables)

import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# Prepare X_test as DataFrame (safe-guard)
if isinstance(X_test, np.ndarray):
    X_test_df = pd.DataFrame(X_test, columns=['TV','Radio','Newspaper'])
else:
    X_test_df = X_test.copy()

# Ensure y_test array
y_test_arr = y_test.values if hasattr(y_test, "values") else np.array(y_test)

# Build predictions (use final_model if available)
if 'final_model' in globals():
    model_to_use = final_model
    scaler_to_use = final_scaler if 'final_scaler' in globals() else None
    if scaler_to_use is not None:
        try:
            X_for_pred = scaler_to_use.transform(X_test_df)
        except Exception:
            X_for_pred = globals().get('X_test_scaled', X_test_df)
    else:
        X_for_pred = X_test_df
    y_pred = model_to_use.predict(X_for_pred)
else:
    # fallback to preds computed earlier
    if 'pred_ridge' in globals():
        y_pred = pred_ridge
    elif 'pred_lr' in globals():
        y_pred = pred_lr
    elif 'pred_lasso' in globals():
        y_pred = pred_lasso
    else:
        raise RuntimeError("No model or predictions found. Run training cells first.")

# Compute RMSE
rmse = np.sqrt(mean_squared_error(y_test_arr, y_pred))

# Helper to add scatter + linear fit
def add_scatter_with_fit(ax, x, y, xlabel, ylabel="Sales"):
    ax.scatter(x, y, alpha=0.75, s=30)
    mask = np.isfinite(x) & np.isfinite(y)
    if mask.sum() > 1:
        coeffs = np.polyfit(x[mask], y[mask], 1)
        xs = np.linspace(np.nanmin(x), np.nanmax(x), 100)
        ax.plot(xs, np.polyval(coeffs, xs), linestyle='--', linewidth=1)
        slope, intercept = coeffs[0], coeffs[1]
        ax.text(0.02, 0.95, f"y={slope:.3f}x+{intercept:.3f}", transform=ax.transAxes,
                fontsize=9, verticalalignment='top', bbox=dict(facecolor='white', alpha=0.6, edgecolor='none'))
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.grid(alpha=0.2)

# Create 2x2 dashboard
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
plt.subplots_adjust(hspace=0.30, wspace=0.28)

# Panel 1: TV vs Sales
add_scatter_with_fit(axes[0,0], X_test_df['TV'].values, y_test_arr, xlabel="TV Spend")
axes[0,0].set_title("TV Spend vs Sales")

# Panel 2: Radio vs Sales
add_scatter_with_fit(axes[0,1], X_test_df['Radio'].values, y_test_arr, xlabel="Radio Spend")
axes[0,1].set_title("Radio Spend vs Sales")

# Panel 3: Newspaper vs Sales
add_scatter_with_fit(axes[1,0], X_test_df['Newspaper'].values, y_test_arr, xlabel="Newspaper Spend")
axes[1,0].set_title("Newspaper Spend vs Sales")

# Panel 4: Predicted vs Actual with identity line
ax = axes[1,1]
ax.scatter(y_pred, y_test_arr, alpha=0.8, s=30)
mn = min(np.nanmin(y_pred), np.nanmin(y_test_arr))
mx = max(np.nanmax(y_pred), np.nanmax(y_test_arr))
ax.plot([mn, mx], [mn, mx], linestyle='--', linewidth=1)
ax.set_xlabel("Predicted Sales")
ax.set_ylabel("Actual Sales")
ax.set_title(f"Predicted vs Actual (RMSE = {rmse:.3f})")
ax.grid(alpha=0.2)

# Annotate top 3 errors on Pred vs Actual
errors = np.abs(y_test_arr - y_pred)
top_idx = np.argsort(errors)[-3:][::-1]
for i in top_idx:
    ax.annotate(f"idx {i}", (y_pred[i], y_test_arr[i]), textcoords="offset points", xytext=(6,4), fontsize=8)

plt.suptitle("Sales Prediction Dashboard  4 Panels", fontsize=16, y=0.98)
plt.show()

# Save image
fig.savefig("sales_dashboard.png", dpi=200, bbox_inches='tight')
print("Saved dashboard to sales_dashboard.png")
```

```python
# ---------- Cell 10: Feature importance + Residuals histogram ----------
# This cell computes a simple feature-importance using model coefficients if available.
# Otherwise it falls back to absolute coef of a linear model (if present).

import matplotlib.pyplot as plt
import numpy as np

feature_names = ['TV','Radio','Newspaper']

# Try to get coefficients from final_model or other models
coef = None
model_source = None
if 'final_model' in globals() and hasattr(final_model, 'coef_'):
    coef = np.asarray(final_model.coef_).ravel()
    model_source = 'final_model'
elif 'best_ridge' in globals() and hasattr(best_ridge, 'coef_'):
    coef = np.asarray(best_ridge.coef_).ravel()
    model_source = 'best_ridge'
elif 'lr' in globals() and hasattr(lr, 'coef_'):
    coef = np.asarray(lr.coef_).ravel()
    model_source = 'lr'

# If we failed to get linear coefficients, make a permutation-style importance (simple)
if coef is None:
    print("No linear coefficients found  using absolute correlation with Sales as proxy for importance.")
    corr_vals = np.array([np.corrcoef(X[feat], y)[0,1] for feat in feature_names])
    importance = np.abs(corr_vals)
else:
    importance = np.abs(coef)  # absolute value of coefficients

# Normalize for visualization
importance_norm = importance / (importance.sum() + 1e-12)

# Residuals for histogram
residuals = y_test_arr - y_pred

# Plot side-by-side: bar (importance) and residuals histogram
fig, axes = plt.subplots(1, 2, figsize=(12,4))
# Importance bar
axes[0].bar(feature_names, importance_norm, alpha=0.85)
axes[0].set_title(f"Feature importance (abs coef)  source: {model_source or 'correlation proxy'}")
axes[0].set_ylabel("Normalized importance")
for i, v in enumerate(importance_norm):
    axes[0].text(i, v + 0.01, f"{v:.2f}", ha='center', fontsize=9)

# Residuals histogram
axes[1].hist(residuals, bins=20, edgecolor='black', alpha=0.7)
axes[1].axvline(0, color='black', linewidth=1)
axes[1].set_title("Residuals distribution (Actual - Predicted)")
axes[1].set_xlabel("Residual")
axes[1].set_ylabel("Count")

plt.suptitle("Feature Importance and Residuals")
plt.tight_layout(rect=[0,0,1,0.95])
plt.show()

# Save
fig.savefig("feature_importance_and_residuals.png", dpi=200, bbox_inches='tight')
print("Saved feature importance chart to feature_importance_and_residuals.png")
```
